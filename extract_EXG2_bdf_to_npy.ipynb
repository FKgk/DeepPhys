```python
import pyedflib
import numpy as np
import matplotlib.pyplot as plt
import os
```

# bdf file
- label: EXG2
- fs: 256
- nsamples: 42496

```python
base = os.path.join(os.getcwd(), 'data')
```

```python
base = os.path.join(os.getcwd(), 'data')
for status in os.listdir(base):
    if status.endswith('csv'):
        continue
        
    for folder in os.listdir(os.path.join(base, status)):
        for file in os.listdir(os.path.join(base, status, folder)):
            if file.endswith('bdf'):
                bdf_data = pyedflib.EdfReader(os.path.join(base, status, folder, file))
                index = bdf_data.getSignalLabels().index('EXG2')
                assert(bdf_data.getLabel(index) == 'EXG2')
                
                signal = bdf_data.readSignal(index)
                bdf_data.close()

                np.save(os.path.join(base, status, folder, "EXG2"), signal)
```

```python
EXG_path = os.path.join(os.getcwd(), 'data', 'train', '10', 'EXG2.npy')
EXG = np.load(EXG_path)
```

```python
fs = 256
start = 0
for i in range(60):
    plt.figure(figsize=(16,8))
    plt.plot(range(fs), EXG[start:start+fs])
    plt.show()
    
    start += fs
```
